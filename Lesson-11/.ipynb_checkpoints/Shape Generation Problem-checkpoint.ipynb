```python
import random
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import PathPatch
from matplotlib.path import Path

from keras.models import Sequential
from keras.layers import LSTM, Dense
```

** Generate Random Rectangles **

```python
# generate a rectangle with random height and width
def random_rectangle():
    width, height = random.random(), random.random()
    points = []
    # bottom
    points.append([0.0, 0.0]) # left
    points.append([width, 0.0]) # right
    # top
    
    points.append([width, height]) # right
    points.append([0.0, height]) # left
    return points
```

```python
rect = random_rectangle()
print(rect)
```

** Plot Rectangles **

```python
# plot a rectangle 
def plot_rectangle(rect):
    # close the rectangle path
    rect.append(rect[0])
    # define path
    codes = [Path.MOVETO, Path.LINETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]
    path = Path(rect, codes)

    axis = plt.gca()
    patch = PathPatch(path)
    # add shape to plot
    axis.add_patch(patch)
    axis.set_xlim(-0.1, 1.1)
    axis.set_ylim(-0.1, 1.1)
    plt.show()
```

```python
rect = random_rectangle()
plot_rectangle(rect)
```

** Rectangle to Sequenc **

```python
# generate input and output sequences for one random rectangle
def get_samples():
    # generate rectangle
    rect = random_rectangle()
    X, y = [], []
    # create input -> output pairs for each coordinate
    for i in range(1, len(rect)):
        X.append(rect[i-1])
        y.append(rect[i])
    # convert input sequence shape to have 1 time step and 2 features
    X, y = np.array(X), np.array(y)
    X = X.reshape((X.shape[0], 1, 2))
    return X, y
```

```python
X, y = get_samples()
for i in range(X.shape[0]):
    print(X[i][0], '=>', y[i])
```

** Define and Compile the Model **

```python
# define model 
model = Sequential()
model.add(LSTM(10, input_shape=(1, 2)))
model.add(Dense(2, activation='linear'))
model.compile(loss='mae', optimizer='adam')
print(model.summary())
```

** Fit the Model **

```python
model_weights = {}
model_snaps = [500, 1000, 5000, 10000, 20000, 25000]
# fit model 
for i in range(25001):
    X, y = get_samples()
    model.fit(X, y, epochs=1, verbose=2, shuffle=False)
    if i in model_snaps:
        model_weights[str(i)] = model.get_weights()
        print('saving model ********************************************************************** \n\n\n\n\n\n')
```

** Make Predictions with the Model **

```python
# use [0, 0] to seed the generatino process
last = np.array([0.0, 0.0]).reshape((1, 1, 2))
# predict the next coordinate
yhat = model.predict(last, verbose=0)
print(yhat)
last = yhat.reshape((1, 1, 2))
yhat = model.predict(last)
print(yhat)
```

```python
# use a fit LSTM to generate new rectangle
def generate_rectangle(model):
    rect = []
    # use [0, 0] to seed the generatino process
    last = np.array([0.0, 0.0]).reshape((1, 1, 2))
    rect.append([[y for y in x] for x in last[0]][0])
    # generate the remainint 3 coordinates
    for i in range(3):
        # predict the next coordinate
        yhat = model.predict(last, verbose=0)
        # use this output as input for the next prediction
        last = yhat.reshape((1, 1, 2))
        # store coordinate
        rect.append([[y for y in x] for x in last[0]][0])
        
    return rect
```

```python
# generate new shapes
rect = generate_rectangle(model)
plot_rectangle(rect)
```

** Evaluate the Model **

```python
len_samples = 3
for snap, model_weight in model_weights.items():
    print('evaluating 3 outputs from snap val %s' %snap)
    model.set_weights(model_weight)
    for i in range(len_samples):
        rect = generate_rectangle(model)
        plot_rectangle(rect)
        plt.show()
```
